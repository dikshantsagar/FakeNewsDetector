# Part 1

###  Binary Classification  as True or False

```python
for i in range(len(train)):
    if(train.iloc[:,1][i]=='true' or train.iloc[:,1][i]=='mostly-true' or train.iloc[:,1][i]=='half-true' ):
        train.iloc[:,1][i]=1
    else:
        train.iloc[:,1][i]=0
```

```python
train=train.replace('\d+', 'NUM', regex=True)
```

```python
cv=CountVectorizer(max_features=10000,ngram_range=(1,3))
```

```python
x=cv.fit_transform(train.iloc[:,0])
```

```python
cv.get_feature_names()
```

```python
cv.vocabulary_
```

```python
clf=MultinomialNB(alpha=4)

clf.fit(x,list(train.iloc[:,1]))

clf.score(x,list(train.iloc[:,1]))
```

```python
clf2=LinearSVC(C=0.1)

clf2.fit(x,list(train.iloc[:,1]))

clf2.score(x,list(train.iloc[:,1]))
```

```python
from sklearn.linear_model import LogisticRegression
cl=LogisticRegression()
cl.fit(x,list(train.iloc[:,1]))
cl.score(x,list(train.iloc[:,1]))
```

```python
from sklearn.naive_bayes import GaussianNB
clf3=GaussianNB()
clf3.fit(x.toarray(),list(train.iloc[:,1]))
clf3.score(x.toarray(),list(train.iloc[:,1]))
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
from keras.preprocessing import text, sequence
from keras import utils

model = Sequential()
model.add(Dense(512, input_shape=(10000,)))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

nn = model.fit(x, utils.to_categorical(train.iloc[:,1], 2),
                    batch_size=30,
                    epochs=20,
                    verbose=1,
                    validation_split=0.1)
```

```python
score = model.evaluate(x,utils.to_categorical(train.iloc[:,1], 2),
                       batch_size=30, verbose=1)
score
```
